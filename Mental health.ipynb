```python
import pandas as pd
data = {
    "User_ID": [1, 2, 3, 4, 5],
    "Age": [36, 34, 65, 34, 22],
    "Gender": ["Non-binary", "Female", "Non-binary", "Male", "Female"],
    "Occupation": ["Sales", "Education", "Sales", "Other", "Healthcare"],
    "Country": ["Canada", "UK", "USA", "Australia", "Canada"],
    "Mental_Health_Condition": ["No", "Yes", "Yes", "No", "Yes"],
    "Severity": ["Medium", None, "High", "Low", "Low"],  # Contains NaN
    "Consultation_History": ["Yes", "No", "No", "No", "No"],
    "Stress_Level": ["Medium", "Low", "Low", "Medium", "Medium"],
    "Sleep_Hours": [7.1, 7.5, 8.4, 9.8, None],  # Contains NaN
    "Work_Hours": [46, 47, 58, 30, 62],
    "Physical_Activity_Hours": [5, 8, 10, 2, 5],
}

df = pd.DataFrame(data)
print("Dataset with missing values:\n", df)
```

```python
df_dropped_na = df.dropna()
print("Dataset after dropping rows with missing values:\n", df_dropped_na)
df_dropped_all_na = df.dropna(how="all")
print("\nDataset after dropping rows where all values are NaN:\n", df_dropped_all_na)
df_dropped_columns = df.dropna(axis=1)
print("\nDataset after dropping columns with missing values:\n", df_dropped_columns)
```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 10, 15, 20, 25]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(x, y1, label='Line Plot 1', color='b', marker='o')
ax1.set_title('First Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

ax2.plot(x, y2, label='Line Plot 2', color='g', marker='s')
ax2.set_title('Second Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
